```python
# 2501 : 약수 구하기

N, K = map(int,input().split())

li = []

for x in range(1, N+1):
  if N%x == 0:
    li.append(x)
  else:
    pass

if len(li) >= K:
  print(li[K-1])
else:
  print(0)
```

```python
# 3460 : 이진수

T = int(input())

for _ in range(T):
    n = bin(int(input()))[2:]
    for i in range(len(n)):
        if n[-i-1] == '1':
            print(i, end = " ")
            

def binary(n):
  i = 0
  while n != 0:
    if n % 2 == 1:
      print(i, end=" ")
    n //= 2
    i += 1

for _ in range(int(input())):
  binary(int(input()))
```

```python
# 10818 : 최소, 최대

N = int(input())
li = list(map(int, input().split()))

print(min(li), max(li), sep=' ')
```

```python
# 2460 : 지능형 기차2

pas_li = []
pas = 0

for i in range(1, 11):
  O, I = map(int, input().split())
  pas += I-O
  pas_li.append(pas)

print(max(pas_li))
```
